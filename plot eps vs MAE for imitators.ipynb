Combine all figs into one plot

```python
import pandas as pd
import numpy as np
```

```python
df_mae = pd.read_csv('default SAC 500 norm space results/imitator (best) fgm eps vs asr/MAE for bin layer DL loss.csv',
                     index_col=0)
df_mae.index = df_mae.index.to_series().round(3)
df_mae.columns = ['500_Training_Episodes_Plus_Bifurcation_Layer',]
```

```python
df_mae.plot.line(ylabel='MAE',
                 xlabel='FGM Epsilons',
                 title='Comparison of MAEs and Perturbation Size for Immitator Snooping Attacks on Different Agents',
                 style=['b-.','b-','r-.','r-'],
                 )
```

```python
df_kpi = pd.read_csv(r'default SAC 500 norm space results/imitator (best) fgm eps vs asr/19 run KPIs bin layer DL loss.csv',
                          index_col=0)
df_kpi.columns = df_kpi.columns.astype(float).to_series().round(3).astype(str)
```

```python
df_kpi
```

```python
electricity_consumption = pd.DataFrame()
electricity_consumption['500_Training_Episodes_Plus_Bifurication_Layer'] = df_kpi.loc['electricity_consumption_total'].transpose()
electricity_consumption.loc['0.00'] = [0.8809147062913397]
electricity_consumption.sort_index(inplace=True)
```

plot KPIs with ASRs/Epsilons

```python
electricity_consumption.plot.line(
    ylabel='Total Electricity Consumption KPI',
    xlabel='FGM Epsilon',
    title='Electricity Consumption vs FGM Epsilon for Immitator Snooping Attack for Different Agents',
    style=['b-.','b-','r-.','r-'],
)
```

```python
cost = pd.DataFrame()
cost['500_Training_Episodes_Plus_Bifurication_Layer'] = df_kpi.loc['cost_total'].transpose()
cost.loc['0.00'] = [0.798446427857767]
cost.sort_index(inplace=True)
```

```python
cost.plot.line(
    ylabel='Total Electricity Cost KPI',
    xlabel='FGM Epsilon',
    title='Electricity Cost vs FGM Epsilon for Immitator Snooping Attack for Different Agents',
    style=['b-.','b-','r-.','r-'],
)
```

```python
plot_kpis = ['electricity_consumption_total','cost_total','daily_peak_average','ramping_average']
```

```python
df_kpi_plot = df_kpi.loc[plot_kpis].transpose()
df_kpi_plot.loc['0.00'] = [0.8809147062913397,0.798446427857767,0.8986220662900906,1.088548505415332]
df_kpi_plot.sort_index(inplace=True)
```

```python
df_kpi_plot.plot.line(
    ylabel='Normalized KPI',
    xlabel='FGM Epsilon',
    title='Electricity Cost and Comsumption vs FGM Epsilon for Immitator Snooping Attack on SAC Agent',
    #style=['b-.','b-','r-.','r-'],
)
```

```python
df_dkpi = pd.read_csv(r'20 bin PPO 500 results\multi eps vs asr for imitator FGM\19 run KPIs bin layer DL loss.csv',
                          index_col=0)
df_dkpi.columns = df_dkpi.columns.astype(float).to_series().round(3).astype(str)
```

```python
df_dkpi_plot = df_dkpi.loc[plot_kpis].transpose()
df_dkpi_plot.loc['0.00'] = [0.8791079206961654,0.7878032417196625,0.8986220662900906,1.0937492364594632]
df_dkpi_plot.sort_index(inplace=True)
```

```python
ax = pd.concat([df_kpi_plot,df_dkpi_plot], axis='columns').plot.line(
    ylabel='Normalized KPI',
    xlabel='FGM Epsilon',
    title='KPIs vs Bifurcated FGM Epsilon for \n Imitator Snooping Attack on Continuous and Discrete Agents',
    style=['b-.','r-.','g-.','c-.','b-','r-','g-','c-'],
    xticks = np.arange(21, step=2),
    figsize=(10,5),
)
ax.legend(['SAC Electricity Consumption',
           'SAC Electricity Cost',
           'SAC Daily Peak',
           'SAC Ramping',
           'Discrete PPO Electricity Consumption',
           'Discrete PPO Electricity Cost',
           'Discrete PPO Daily Peak',
           'Discrete PPO Ramping',
           ])
```
