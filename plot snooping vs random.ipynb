```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pylab as pylab

params = {'axes.labelsize': 'x-large',
          'xtick.labelsize': 'large',
          'ytick.labelsize': 'large',
          'axes.titlesize': 'x-large',
          #'legend.fontsize': 'x-large',
          #'font.size': 16,
          'axes.titlesize': 'x-large',
          'figure.titlesize': 'x-large',}
pylab.rcParams.update(params)
```

```python
df_asr = pd.read_csv('20 bin PPO 500 results\multi eps vs asr for imitator FGM\ASR for varied epsilons 1.csv',
                     index_col=0)

df_asr['ATLA_ASRs'] = pd.read_csv('PPO agent 100 alts over 1000+200 2-3-21 results\imitator fgm eps vs asr\ASR for varied epsilons.csv',
                                 index_col=0)

df_asr['rand_ASRs'] = pd.read_csv(r'20 bin PPO 500 results\random noise eps vs ASR results\ASR for varied epsilons 1.csv',
                                 index_col=0)

df_asr['rand_ATLA_ASRs'] = pd.read_csv(r'PPO agent 100 alts over 1000+200 2-3-21 results\random noise eps vs ASR results\ASR for varied epsilons 1.csv',
                                 index_col=0)
```

```python
df_asr.index = df_asr.index.to_series().round(3)
df_asr.columns = ['Snooping_500_Training_Episodes','Snooping_500_Training_Episodes_Plus_ATLA', 'Random_500_Training_Episodes', 'Random_500_Training_Episodes_ATLA']
```

```python
df_asr.plot.line(ylabel='ASR',
                 xlabel='Epsilons',
                 title='Comparison of ASRs and Perturbation Size for Imitator\nSnooping Attacks and Random Noise on Different Agents',
                 style=['-.','-','-.','-'],
                 )
```

```python
df_atla_kpi = pd.read_csv(r'PPO agent 100 alts over 1000+200 2-3-21 results\imitator fgm eps vs asr\19 run KPIs 1.csv',
                          index_col=0)
df_atla_kpi.columns = df_atla_kpi.columns.astype(float).to_series().round(3).astype(str)
```

```python
df_kpi = pd.read_csv(r'20 bin PPO 500 results\multi eps vs asr for imitator FGM\19 run KPIs 1.csv',
                          index_col=0)
df_kpi.columns = df_kpi.columns.astype(float).to_series().round(3).astype(str)
```

```python
df_bin_kpi = pd.read_csv(r'20 bin PPO 500 results\random noise eps vs ASR results\19 run KPIs 1.csv',
                          index_col=0)
df_bin_kpi.columns = df_bin_kpi.columns.astype(float).to_series().round(3).astype(str)
```

```python
df_bin_atla_kpi = pd.read_csv(r'PPO agent 100 alts over 1000+200 2-3-21 results\random noise eps vs ASR results\19 run KPIs 1.csv',
                          index_col=0)
df_bin_atla_kpi.columns = df_bin_atla_kpi.columns.astype(float).to_series().round(3).astype(str)
```

```python
df_kpi
```

```python
electricity_consumption = pd.DataFrame()
electricity_consumption['Snooping_500_Training_Episodes'] = df_kpi.loc['electricity_consumption_total'].transpose()
electricity_consumption['Snooping_500_Training_Episodes_Plus_ATLA'] = df_atla_kpi.loc['electricity_consumption_total'].transpose()
electricity_consumption['Random_500_Training_Episodes'] = df_bin_kpi.loc['electricity_consumption_total'].transpose()
electricity_consumption['Random_500_Training_Episodes_Plus_ATLA'] = df_bin_atla_kpi.loc['electricity_consumption_total'].transpose()
```

```python
electricity_consumption.index
```

```python
electricity_consumption.loc['0.00'] = [0.8791079206961654,0.907584246758169,0.8791079206961654,0.907584246758169]
electricity_consumption.sort_index(inplace=True)
```

plot KPIs with ASRs/Epsilons

```python
electricity_consumption.plot.line(
    ylabel='Total Electricity Consumption KPI',
    xlabel='FGM Epsilon',
    title='Electircity Consumption vs FGM Epsilon for Immitator Snooping Attack for Different Agents',
    style=['-.','-','-.','-']
)
```

NB that for the graph above an epsilon of 0 indicates no attack

```python
(electricity_consumption/electricity_consumption.loc['0.00']).plot.line(
    ylabel='Normalized Total Electricity Consumption KPI',
    xlabel='FGM Epsilon',
    title='Electircity Consumption vs FGM Epsilon for Immitator Snooping Attack',
    style=['-.','-','-.','-']
    )
```

KPIs are normalized by values for clean episodes. We see a much smaller increase from the ATLA agent
