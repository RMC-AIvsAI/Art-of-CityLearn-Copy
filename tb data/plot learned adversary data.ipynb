```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import matplotlib.pylab as pylab

params = {'axes.labelsize': 'x-large',
          'xtick.labelsize': 'x-large',
          'ytick.labelsize': 'x-large',
          'axes.titlesize': 'x-large',
          'legend.fontsize': 'x-large'}
pylab.rcParams.update(params)
```

```python
df_eval1 = pd.read_csv('ATLA_SAC_RWD3_1 eval.csv', usecols=['Value'])
df_eval1.columns = ['evaluation_scores']
df_eval2 = pd.read_csv('ATLA_SAC_RWD3_2 eval.csv', usecols=['Value'])
df_eval2.columns = ['evaluation_scores']
df_eval3 = pd.read_csv('ATLA_SAC_RWD3_300_2  eval.csv', usecols=['Value'])
df_eval3.columns = ['evaluation_scores']
```

```python
plt.figure(figsize=(10,3))
#plt.plot(df_eval1, label='SAC 300 Training Episodes')
plt.plot(df_eval2, label='SAC 600 Training Episodes')
#plt.plot(df_eval3, label='3')

plt.xlabel('Evaluation number (once every 10 training episodes)')
plt.ylabel('Evaluation Reward')
plt.title('Learned Adversary (LA) Evaluation Rewards')
plt.legend()
```

```python
plt.figure(figsize=(20,6))
plt.plot(df_eval1, label='SAC 300 Training Episodes')
plt.plot(df_eval2, label='SAC 600 Training Episodes')
#plt.plot(df_eval3, label='3')

plt.xlabel('Evaluation number (once every 10 training episodes)')
plt.ylabel('Evaluation Reward')
plt.title('Learned Adversary (LA) Evaluation Rewards')
plt.legend()
```

```python
df_size1 = pd.read_csv('ATLA_SAC_RWD3_1 pertubation size.csv', usecols=['Value'])
df_size1.columns = ['perturbation_size']
df_size2 = pd.read_csv('ATLA_SAC_RWD3_2 pertubation size.csv', usecols=['Value'])
df_size2.columns = ['perturbation_size']
df_size3 = pd.read_csv('ATLA_SAC_RWD3_300_2 pertubation size.csv', usecols=['Value'])
df_size3.columns = ['perturbation_size']
```

```python
df_size3.count()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df_size2, color='blue', label='Before restart')

# Plot the scores after the restart, continuing from where the previous scores ended
plt.plot(range(len(df_size2), len(df_size2) + len(df_size3)), df_size3, color='red', label='After restart')

plt.xlabel('Training Episodes')
plt.ylabel('Linf Perturbation distance')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20,6))
plt.plot(df_size1, label='SAC 300 Training Episodes')
plt.plot(df_size2, label='SAC 600 Training Episodes')
#plt.plot(df_size3, label='3')

plt.xlabel('Episode')
plt.ylabel('Episode mean L infinity distance from original')
plt.title('Learned Adversary (LA) Mean Perturbation Size over Training Episodes')
plt.legend()
```
