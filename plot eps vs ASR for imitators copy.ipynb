Combine all figs into one plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
import matplotlib.pylab as pylab

params = {'axes.labelsize': 'x-large',
          'xtick.labelsize': 'x-large',
          'ytick.labelsize': 'x-large',
          'axes.titlesize': 'x-large',
          'legend.fontsize': 'x-large',
          #'font.size': 16,
          'axes.titlesize': 'x-large'}
pylab.rcParams.update(params)
```

```python
df_ppod = pd.read_csv(r'20 bin PPO 500 results\multi eps vs asr for imitator FGM\19 run KPIs 1.csv',
                          index_col=0)
df_ppod.columns = df_ppod.columns.astype(float).to_series().round(3).astype(str)
```

```python
df_ppoc = pd.read_csv(r'Karla\3-26-5 PPOc Karla results\imitator (best) fgm eps vs asr\19 run KPIs bifurcation DL loss.csv',
                          index_col=0)
df_ppoc.columns = df_ppoc.columns.astype(float).to_series().round(3).astype(str)
```

```python
df_sac = pd.read_csv(r'default SAC 500 norm space results\imitator (best) fgm eps vs asr\19 run KPIs bin layer DL loss.csv',
                          index_col=0)
df_sac.columns = df_sac.columns.astype(float).to_series().round(3).astype(str)
```

```python
df_ppoc.index
```

```python
electricity_consumption = pd.DataFrame()
electricity_consumption['Discrete_PPO'] = df_ppod.loc['electricity_consumption_total'].transpose()
electricity_consumption['Continuous_PPO'] = df_ppoc.loc['electricity_consumption_total'].transpose()
electricity_consumption['SAC'] = df_sac.loc['electricity_consumption_total'].transpose()
```

```python
electricity_consumption.loc['0.00'] = [0.8791079206961654,0.8743575386801745,0.8809147062913397]
electricity_consumption.sort_index(inplace=True)
```

```python
cost = pd.DataFrame()
cost['Discrete_PPO'] = df_ppod.loc['cost_total'].transpose()
cost['Continuous_PPO'] = df_ppoc.loc['cost_total'].transpose()
cost['SAC'] = df_sac.loc['cost_total'].transpose()
cost.loc['0.00'] = [0.7878032417196625,0.8137059576458006,0.798446427857767]
cost.sort_index(inplace=True)
```

```python
peak = pd.DataFrame()
peak['Discrete_PPO'] = df_ppod.loc['daily_peak_average'].transpose()
peak['Continuous_PPO'] = df_ppoc.loc['daily_peak_average'].transpose()
peak['SAC'] = df_sac.loc['daily_peak_average'].transpose()
peak.loc['0.00'] = [0.8986220662900906,0.8914214653074561,0.8986220662900906]
peak.sort_index(inplace=True)
ramp = pd.DataFrame()
ramp['Discrete_PPO'] = df_ppod.loc['ramping_average'].transpose()
ramp['Continuous_PPO'] = df_ppoc.loc['ramping_average'].transpose()
ramp['SAC'] = df_sac.loc['ramping_average'].transpose()
ramp.loc['0.00'] = [1.0937492364594632,0.9426702276964324,1.088548505415332]
ramp.sort_index(inplace=True)
```

Rearrange as column

```python
x_ticks = np.arange(21, step=2)
row = 4
col = 1
fig = plt.figure(figsize=[8,18],)
fig.suptitle('KPIs for Imitator Bifurcated Snooping Attack for Different Agents\n',
             fontsize='x-large')
plt.subplot(row, col, 1) #(row, col, plt#)
electricity_consumption.plot.line(
    ylabel='Total Electricity Consumption KPI',
    xlabel='FGM Epsilon',
    title='Electricity Consumption vs FGM Epsilon',
    #style=['b-.','b-','r-.','r-'],
    xticks=x_ticks,
    ax=plt.gca(), #get current axis
)
plt.subplot(row, col, 2)
cost.plot.line(
    ylabel='Total Electricity Cost KPI',
    xlabel='FGM Epsilon',
    title='Electricity Cost vs FGM Epsilon',
    #style=['b-.','b-','r-.','r-'],
    xticks=x_ticks,
    ax=plt.gca(), #get current axis
)
plt.subplot(row, col, 3) #(row, col, plt#)
peak.plot.line(
    ylabel='Daily Peak Average KPI',
    xlabel='FGM Epsilon',
    title='Daily Peak Average vs FGM Epsilon',
    #style=['b-.','b-','r-.','r-'],
    xticks=x_ticks,
    ax=plt.gca(), #get current axis
)
plt.subplot(row, col, 4)
ramp.plot.line(
    ylabel='Ramping Average KPI',
    xlabel='FGM Epsilon',
    title='Ramping Average vs FGM Epsilon',
    #style=['b-.','b-','r-.','r-'],
    xticks=x_ticks,
    ax=plt.gca(), #get current axis
)
plt.tight_layout()
```
