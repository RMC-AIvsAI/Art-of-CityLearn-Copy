```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
kwargs = dict(
    header=0, 
    usecols=['baseline','cost_function'], #data, index
    index_col=0 
)
```

```python
df = pd.read_csv('PPO agent 20 alts over 250 1-9-21 results\KPIs.csv', **kwargs)
df.columns = ['1_9_21']
```

```python
df = df.merge(pd.read_csv('PPO agent 30 alts over 0+450 1-11-14 results\KPIs.csv', **kwargs),
         left_index=True,
         right_index=True,
         how='outer',
         ).rename(columns={'baseline':'1_11_14'})
```

```python
df = df.merge(pd.read_csv('PPO agent 20 alts over 0+200 1-12-21 results\KPIs.csv', **kwargs),
         left_index=True,
         right_index=True,
         how='outer',
         ).rename(columns={'baseline':'1_12_21'})
```

```python
df = df.merge(pd.read_csv('PPO agent 20 alts over 0+200 1-14-15 results\KPIs.csv', **kwargs),
         left_index=True,
         right_index=True,
         how='outer',
         ).rename(columns={'baseline':'1_14_15'})
```

```python
df = df.merge(pd.read_csv('PPO agent 100 alts over 0+500 1-17-21 results\KPIs.csv', **kwargs),
         left_index=True,
         right_index=True,
         how='outer',
         ).rename(columns={'baseline':'1_17_21'})
```

```python
df = df.merge(pd.read_csv('PPO agent 100 alts over 500+500 1-30-15 results\KPIs.csv', **kwargs),
         left_index=True,
         right_index=True,
         how='outer',
         ).rename(columns={'baseline':'1_30_15'})
```

```python
df = df.merge(pd.read_csv('PPO agent 100 alts over 1000+200 2-3-21 results\KPIs.csv', **kwargs),
         left_index=True,
         right_index=True,
         how='outer',
         ).rename(columns={'baseline':'1_3_21'})
```

```python
key_KPIs = ['electricity_consumption_total','ramping_average','daily_peak_average','cost_total']
names = ['20 Episode Pre-Training, 10 Alternations',
         '300 Episode Pre-Training, 15 Alternations',
         'Reduced B(s), 300 Episode Pre-Training,\n10 Alternations',
         'Reduced B(s)/2, 500 Episode Pre-Training,\n10 Alternations',
         'Reduced B(s), 500 Episode Pre-Training,\n10 Alternations of 100 Episodes',
         'Reduced B(s), 500 Episode Pre-Training,\n10 Alternations of 100 Episodes,\nAgent Trains First',
         'Continued from Best']
ax = df.loc[key_KPIs].plot.barh(
    xlim=0.8,
    figsize=(8,8),
    title='KPIs for Agents Trained with Varied ATLA Hyperparameters'
)
for container in ax.containers:
    ax.bar_label(container,fmt='%.3f')
ax.legend(names,
    bbox_to_anchor=(0.45, 1.0),
    )
ax.set_xlabel('Normalized KPI')
```
