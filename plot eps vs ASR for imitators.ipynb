```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
import matplotlib.pylab as pylab

params = {'axes.labelsize': 'x-large',
          'xtick.labelsize': 'x-large',
          'ytick.labelsize': 'x-large',
          'axes.titlesize': 'x-large',
          'legend.fontsize': 'large',
          #'font.size': 16,
          'axes.titlesize': 'x-large'}
pylab.rcParams.update(params)
```

```python
df_asr = pd.read_csv('20 bin PPO 500 results\multi eps vs asr for imitator FGM\ASR for varied epsilons 1.csv',
                     index_col=0)

df_asr['ATLA_ASRs'] = pd.read_csv('PPO agent 100 alts over 1000+200 2-3-21 results\imitator fgm eps vs asr\ASR for varied epsilons.csv',
                                 index_col=0)

df_asr['Bin_ASRs'] = pd.read_csv('20 bin PPO 500 results\multi eps vs asr for imitator FGM\ASR for varied epsilons bin layer DL loss.csv',
                                 index_col=0)

df_asr['Bin_ATLA_ASRs'] = pd.read_csv('PPO agent 100 alts over 1000+200 2-3-21 results\imitator fgm eps vs asr\ASR for varied epsilons bin layer DL loss.csv',
                                 index_col=0)
```

```python
df_asr.index = df_asr.index.to_series().round(3)
df_asr.columns = ['500_Training_Episodes','500_Training_Episodes_Plus_ATLA', '500_Training_Episodes_Plus_Bifurication_Layer', '500_Training_Episodes_Plus_Bifurication_Layer_and_ATLA']
```

```python
df_asr.plot.line(ylabel='ASR',
                 xlabel='FGM Epsilons',
                 title='Comparison of ASRs and Perturbation Size for Immitator Snooping Attacks on Different Agents',
                 style=['b-.','b-','r-.','r-'],
                 )
```

```python
df_atla_kpi = pd.read_csv(r'PPO agent 100 alts over 1000+200 2-3-21 results\imitator fgm eps vs asr\19 run KPIs 1.csv',
                          index_col=0)
df_atla_kpi.columns = df_atla_kpi.columns.astype(float).to_series().round(3).astype(str)
```

```python
df_kpi = pd.read_csv(r'20 bin PPO 500 results\multi eps vs asr for imitator FGM\19 run KPIs 1.csv',
                          index_col=0)
df_kpi.columns = df_kpi.columns.astype(float).to_series().round(3).astype(str)
```

```python
df_bin_kpi = pd.read_csv(r'20 bin PPO 500 results\multi eps vs asr for imitator FGM\19 run KPIs bin layer DL loss.csv',
                          index_col=0)
df_bin_kpi.columns = df_bin_kpi.columns.astype(float).to_series().round(3).astype(str)
```

```python
df_bin_atla_kpi = pd.read_csv(r'PPO agent 100 alts over 1000+200 2-3-21 results\imitator fgm eps vs asr\19 run KPIs bin layer DL loss.csv',
                          index_col=0)
df_bin_atla_kpi.columns = df_bin_atla_kpi.columns.astype(float).to_series().round(3).astype(str)
```

```python
df_kpi
```

```python
electricity_consumption = pd.DataFrame()
electricity_consumption['500_Training_Episodes'] = df_kpi.loc['electricity_consumption_total'].transpose()
electricity_consumption['500_Training_Episodes_Plus_ATLA'] = df_atla_kpi.loc['electricity_consumption_total'].transpose()
electricity_consumption['500_Training_Episodes_Plus_Bifurication_Layer'] = df_bin_kpi.loc['electricity_consumption_total'].transpose()
electricity_consumption['500_Training_Episodes_Plus_Bifurication_Layer_and_ATLA'] = df_bin_atla_kpi.loc['electricity_consumption_total'].transpose()
```

```python
electricity_consumption.loc['0.00'] = [0.8791079206961654,0.907584246758169,0.8791079206961654,0.907584246758169]
electricity_consumption.sort_index(inplace=True)
```

plot KPIs with ASRs/Epsilons

```python
electricity_consumption.plot.line(
    ylabel='Total Electricity Consumption KPI',
    xlabel='FGM Epsilon',
    title='Electricity Consumption vs FGM Epsilon for Imitator\nSnooping Attack for Different Discrete PPO Agents',
    style=['b-.','b-','r-.','r-'],
    xticks=np.arange(21, step=2),
)
plt.legend(prop={'size': 10})
```

##### ATLA increase resistance to the bifircation attack!

NB that for the graph above an epsion of 0 indicates no attack

```python
(electricity_consumption/electricity_consumption.loc['0.00']).plot.line(
    ylabel='Normalized Total Electricity Consumption KPI',
    xlabel='FGM Epsilon',
    title='Electricity Consumption vs FGM Epsilon for Imitator Snooping Attack',
    style=['b-.','b-','r-.','r-'],
    )
```

KPIs are normalized by values for clean episodes. We see a much smaller increase from the ATLA agent

```python
cost = pd.DataFrame()
cost['500_Training_Episodes'] = df_kpi.loc['cost_total'].transpose()
cost['500_Training_Episodes_Plus_ATLA'] = df_atla_kpi.loc['cost_total'].transpose()
cost['500_Training_Episodes_Plus_Bifurication_Layer'] = df_bin_kpi.loc['cost_total'].transpose()
cost['500_Training_Episodes_Plus_Bifurication_Layer_and_ATLA'] = df_bin_atla_kpi.loc['cost_total'].transpose()
cost.loc['0.00'] = [0.7878032417196625,0.80,0.7878032417196625,0.80]
cost.sort_index(inplace=True)
```

```python
cost.plot.line(
    ylabel='Total Electricity Cost KPI',
    xlabel='FGM Epsilon',
    title='Electricity Cost vs FGM Epsilon for Imitator Snooping Attack for Different Agents',
    style=['b-.','b-','r-.','r-'],
)
```

```python
(cost/cost.loc['0.00']).plot.line(
    ylabel='Normalized Total Electricity Cost KPI',
    xlabel='FGM Epsilon',
    title='Electricity Cost vs FGM Epsilon for Imitator Snooping Attack',
    style=['b-.','b-','r-.','r-'],
    )
```

```python
fig = plt.figure(figsize=[20,5],)
fig.suptitle('KPIs for Imitator Snooping Attack for Different Discrete PPO Agents')
plt.subplot(1, 2, 1) #(row, col, plt#)
electricity_consumption.plot.line(
    ylabel='Total Electricity Consumption KPI',
    xlabel='FGM Epsilon',
    title='Electricity Consumption vs FGM Epsilon',
    style=['b-.','b-','r-.','r-'],
    ax=plt.gca(), #get current axis
)
plt.subplot(1, 2, 2)
cost.plot.line(
    ylabel='Total Electricity Cost KPI',
    xlabel='FGM Epsilon',
    title='Electricity Cost vs FGM Epsilon',
    style=['b-.','b-','r-.','r-'],
    ax=plt.gca(), #get current axis
)
```

```python
peak = pd.DataFrame()
peak['500_Training_Episodes'] = df_kpi.loc['daily_peak_average'].transpose()
peak['500_Training_Episodes_Plus_ATLA'] = df_atla_kpi.loc['daily_peak_average'].transpose()
peak['500_Training_Episodes_Plus_Bifurication_Layer'] = df_bin_kpi.loc['daily_peak_average'].transpose()
peak['500_Training_Episodes_Plus_Bifurication_Layer_and_ATLA'] = df_bin_atla_kpi.loc['daily_peak_average'].transpose()
peak.loc['0.00'] = [0.8986220662900906,0.91796042517278,0.8986220662900906,0.91796042517278]
peak.sort_index(inplace=True)
ramp = pd.DataFrame()
ramp['500_Training_Episodes'] = df_kpi.loc['ramping_average'].transpose()
ramp['500_Training_Episodes_Plus_ATLA'] = df_atla_kpi.loc['ramping_average'].transpose()
ramp['500_Training_Episodes_Plus_Bifurication_Layer'] = df_bin_kpi.loc['ramping_average'].transpose()
ramp['500_Training_Episodes_Plus_Bifurication_Layer_and_ATLA'] = df_bin_atla_kpi.loc['ramping_average'].transpose()
ramp.loc['0.00'] = [1.0937492364594632,1.1178860472017884,1.0937492364594632,1.1178860472017884]
ramp.sort_index(inplace=True)
```
