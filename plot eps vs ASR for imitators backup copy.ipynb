Combine all figs into one plot

```python
import pandas as pd
import numpy as np
```

```python
df_asr = pd.read_csv('20 bin PPO 500 results\multi eps vs asr for imitator FGM\ASR for varied epsilons.csv',
                     index_col=0)

df_asr['ATLA_ASRs'] = pd.read_csv('PPO agent 100 alts over 1000+200 2-3-21 results\imitator fgm eps vs asr\ASR for varied epsilons.csv',
                                 index_col=0)
```

```python
df_asr.index = df_asr.index.to_series().round(3)
df_asr.columns = ['500_Training_Episodes','500_Training_Episodes_Plus_ATLA']
```

```python
df_asr.plot.line(ylabel='ASR',
                 xlabel='FGM Epsilons',
                 title='Comparison of ASRs and Perturbation Size for Immitator Snooping Attacks on Different Agents',
                 )
```

```python
df_atla_kpi = pd.read_csv(r'PPO agent 100 alts over 1000+200 2-3-21 results\imitator fgm eps vs asr\19 run KPIs 1.csv',
                          index_col=0)
df_atla_kpi.columns = df_atla_kpi.columns.astype(float).to_series().round(3).astype(str)
```

```python
df_kpi = pd.read_csv(r'20 bin PPO 500 results\multi eps vs asr for imitator FGM\19 run KPIs 1.csv',
                          index_col=0)
df_kpi.columns = df_kpi.columns.astype(float).to_series().round(3).astype(str)
```

```python
df_kpi
```

```python
electricity_consumption = pd.DataFrame()
electricity_consumption['500_Training_Episodes'] = df_kpi.loc['electricity_consumption_total'].transpose()
electricity_consumption['500_Training_Episodes_Plus_ATLA'] = df_atla_kpi.loc['electricity_consumption_total'].transpose()
```

```python
electricity_consumption.index
```

```python
electricity_consumption.loc['0.00'] = [0.8791079206961654,0.907584246758169]
electricity_consumption.sort_index(inplace=True)
```

plot KPIs with ASRs/Epsilons

```python
electricity_consumption.plot.line(
    ylabel='Total Electricity Consumption KPI',
    xlabel='FGM Epsilon',
    title='Electircity Consumption vs FGM Epsilon for Immitator Snooping Attack for Different Agents',
)
```

NB that for the graph above an epsion of 0 indicates no attack

```python
(electricity_consumption/electricity_consumption.loc['0.00']).plot.line(
    ylabel='Normalized Total Electricity Consumption KPI',
    xlabel='FGM Epsilon',
    title='Electircity Consumption vs FGM Epsilon for Immitator Snooping Attack',
    )
```

KPIs are normalized by values for clean episodes. We see a much smaller increase from the ATLA agent
