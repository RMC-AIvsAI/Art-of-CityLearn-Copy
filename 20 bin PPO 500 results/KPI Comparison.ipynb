```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df_kpis = pd.read_csv('KPIs.csv', index_col='cost_function')
```

```python
df_kpis[['baseline','random_noise','untargeted_acg','untargeted_bb']].plot.barh(xlim=0.7, title='KPI Comparison of Untargeted Attacks').legend(bbox_to_anchor=(1.0, 1.05))
```

```python
df_kpis[['baseline','optimal_acg','optimal_myBB']].plot.barh(xlim=0.7, title='KPI Comparison of Optimal Attacks').legend(bbox_to_anchor=(1.0, 1.05))
```

```python
df_kpis[['baseline','untargeted_bb','optimal_myBB']].plot.barh(xlim=0.7, 
                                                               title='KPI Comparison of Best Optimal and Untargeted Attacks',
                                                               ).legend(bbox_to_anchor=(1.0, 1.05))
```

```python
df_kpis.columns
```

```python
kpi_to_plot = ['cost_total','daily_peak_average','electricity_consumption_total','ramping_average']
scenarios = ['baseline','optimal_myBB','untargeted_acg','untargeted_bb']
ax = df_kpis[scenarios].loc[kpi_to_plot].plot.barh(
    title='KPI Comparison of Best Optimal and Untargeted Attacks',
    figsize = (7.5,5)
                                                               )
for container in ax.containers:
    ax.bar_label(container,fmt='%.2f')
ax.legend(['Clean','Optimally Targeted BB','Dynamic Untargeted ACG','Untargeted BB'])
ax.set_xlabel('Normalized KPI')
plt.tight_layout()
```

```python
kpi_to_plot = ['cost_total','daily_peak_average','electricity_consumption_total','ramping_average']
scenarios = ['baseline','untargeted_acg']
ax = df_kpis[scenarios].loc[kpi_to_plot].plot.barh(
    title='Untargeted ACG attack KPIs Compared with Clean Performance',
    figsize = (8,3),
    xlim=0.7,
                                                               )
for container in ax.containers:
    ax.bar_label(container,fmt='%.2f')
ax.legend(['Clean','Untargeted ACG'])
ax.set_xlabel('Normalized KPI')
plt.tight_layout()
```

```python
kpi_to_plot = ['cost_total','daily_peak_average','electricity_consumption_total','ramping_average']
scenarios = ['baseline','untargeted_bb']
ax = df_kpis[scenarios].loc[kpi_to_plot].plot.barh(
    title='Untargeted BB attack KPIs Compared with Clean Performance',
    figsize = (8,3),
    xlim=0.7,
                                                               )
for container in ax.containers:
    ax.bar_label(container,fmt='%.2f')
ax.legend(['Clean','Untargeted BB'])
ax.set_xlabel('Normalized KPI')
plt.tight_layout()
```

```python
kpi_to_plot = ['cost_total','daily_peak_average','electricity_consumption_total','ramping_average']
scenarios = ['untargeted_acg','untargeted_bb']
ax = df_kpis[scenarios].sub(df_kpis['baseline'].values, axis='rows').loc[kpi_to_plot].plot.barh(
    title='Adversarial Regrets for Untargeted ACG and BB Attacks',
    figsize=(8,3),
    )
for container in ax.containers:
    ax.bar_label(container,fmt='%.2f')
ax.legend(['Untargeted ACG','Untargeted BB'])
ax.set_xlabel('Normalized KPI Difference from Clean Values')
plt.tight_layout()
```
