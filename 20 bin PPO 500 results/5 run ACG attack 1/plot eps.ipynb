```python
import pandas as pd
```

```python
df_eps = pd.read_csv("5 run epsilons 1.csv",
                     index_col=0,
                     header=0)
```

```python
df_eps
```

```python
eps_value_counts = df_eps.apply(lambda col: col.value_counts(), axis='rows')
ASRs = df_eps.notna().mean()
ASRs.name = 'ASR'
eps_value_counts = eps_value_counts.append(ASRs)
eps_value_counts[['mean', 'std', 'variance']] = eps_value_counts.agg(['mean','std', 'var'], axis='columns')
```

```python
eps_value_counts
```

```python
eps_value_counts['mean'].iloc[:-1].plot.pie(autopct='%1.1f%%', title=f'Mean proportions of different $\epsilon$ values over 5 runs of the untargeted ACG attack using the dynamic distortion')
```
