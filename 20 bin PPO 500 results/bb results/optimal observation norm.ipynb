```python
import numpy as np
import pandas as pd
```

Norms for the untargeted bb attack

```python
df_clean = pd.read_csv("optimal BB observations.csv")
df_adv = pd.read_csv("optimal BB adversarial observations.csv")
```

```python
adv_norms = pd.Series(np.linalg.norm(df_adv - df_clean, ord=np.inf, axis=1))
```

```python
adv_norms.describe()
```

try setting xlim to 0

```python
adv_norms.plot.kde(grid=True, 
                   xticks=np.arange(start=adv_norms.min(), stop=adv_norms.max(), step=0.05),
                   figsize=(20,12),
                   xlim=[adv_norms.min(),adv_norms.max()],
                   title='KDE of Linf Norms for Untargeted Advsersarial Observations Generated by BB Attack',
                   )
```

```python
adv_norms.plot.hist(bins=100,
                    xticks=np.arange(start=adv_norms.min(), stop=adv_norms.max(), step=0.025),
                    figsize=(20,12),
                    )
```
