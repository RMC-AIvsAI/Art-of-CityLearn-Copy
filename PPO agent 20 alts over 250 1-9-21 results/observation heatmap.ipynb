```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df_acg_clean = pd.read_csv("dynamic_untargeted_acg_obs.csv",
                       index_col=0)
df_acg_adv = pd.read_csv("dynamic_untargeted_acg_adv_obs.csv",
                     index_col=0)
```

```python
df_bb_clean = pd.read_csv("untargeted_mybb_obs.csv",
                       index_col=0)
df_bb_adv = pd.read_csv("untargeted_mybb_adv_obs.csv",
                     index_col=0)
```

```python

def plot_rows(df1, df2, row_ids):
    fig, axs = plt.subplots(1, 3, 
                            figsize=(25, 35),
                            sharex=True, 
                            sharey=True,
                            )
    df_diff = (df1 - df2).abs()

    # Plot row from df1
    sns.heatmap(df1.iloc[row_ids], 
                ax=axs[0], 
                cmap='viridis',
                cbar=False)
    axs[0].set_title('Clean Observations')

    # Plot row from df2
    sns.heatmap(df2.iloc[row_ids], 
                ax=axs[1], 
                cmap='viridis',
                cbar=False
                )
    axs[1].set_title('Adversarial Observations')

    # Plot the difference
    sns.heatmap(df_diff.iloc[row_ids], 
                ax=axs[2], 
                cmap='viridis',
                vmin=0, 
                vmax=1,
                )
    axs[2].set_title('Absolute Difference')

    fig.text(0.5, 0.001, 'Features', ha='center')
    fig.text(0.1, 0.5, 'Time Step', va='center', rotation='vertical')
    fig.suptitle('Comparison of Clean and Adversarial Observations')
    plt.show()
```

```python
plot_rows(df_bb_clean, df_bb_adv, range(0,100))
```

```python
plot_rows(df_acg_clean, df_acg_adv, range(0,100))
```
