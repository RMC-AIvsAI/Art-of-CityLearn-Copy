```python
df_kpis = pd.read_csv('reward_comparison_KPIs.csv', index_col=0)
```

```python
display(df_kpis)
```

```python
rwd_names = ['Cost-SoC','Energy-SoC','Solar','Cost','Energy']
```

```python
ax = df_kpis.plot.barh(xlim=0.7,
                       title='PPO KPIs for Varied Reward Functions')
ax.legend(rwd_names)
ax.set_xlabel('Normalized KPI')
```

```python
metrics = ['electricity_consumption_total','daily_peak_average','ramping_average', 'cost_total']
ax = df_kpis.loc[metrics].plot.barh(xlim=0.7,title='Select PPO KPIs for Varied Reward Functions')
for container in ax.containers:
    ax.bar_label(container,fmt='%.3f')
ax.legend(rwd_names)
ax.set_xlabel('Normalized KPI')
```
